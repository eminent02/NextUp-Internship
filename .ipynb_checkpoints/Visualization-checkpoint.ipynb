```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
original_df = pd.read_csv('creditcard.csv')
original_df.head()
```

```python
original_df.shape
```

```python
df = pd.read_csv('creditcard_cleaned.csv')
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df.shape
```

### Graphs for original (uncleaned) data

```python
sns.countplot('Class', data=original_df, palette=["green", "red"])
plt.title('Class Distributions \n (No Fraud : Fraud)', fontsize=14)
```

Distribution is skewed as displayed in the above graph

We shall see how the features - time and amount are distributed

```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = original_df['Amount'].values
time_val = original_df['Time'].values

sns.distplot(amount_val, ax=ax[0], color='brown')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(amount_val), max(amount_val)])
ax[0].set_xlabel('Transaction amount')
ax[0].set_ylabel('Frequency')

sns.distplot(time_val, ax=ax[1], color='orange')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(time_val), max(time_val)])
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Likelihood')
plt.show()
```

From the above graphs we observe that :
1. The transaction amount distributionn - we observe that small amount of transcation are in majority while higher amount of transactions are rare.
2. The transaction time distribution - we observe that during the time (0 to 20,000) and (90,000 to 110,000) the likelihood is lesser than 20% and other times the likelihood is closer to 80%

## Correlation Matrix for the original data

```python
corr = original_df.corr()
plt.figure(figsize=(15,8))
sns.heatmap(corr, cmap='coolwarm_r', annot = False)
plt.title("Imbalanced Correlation Matrix \n (don't use for reference)", fontsize=14)
```

Conclusion - Original data does not help us find corelations

## Plotting graphs for cleaned data

```python
print('Distribution of the Classes in the subsample dataset')
print(df['Class'].value_counts()/len(df))
sns.countplot('Class', data=df, palette=["green", "red"])
plt.title('Equally Distributed Classes', fontsize=14)
plt.show()
```

The cleaned dataset has equally distributed classes

```python
corr = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm_r', annot = False)
plt.title("SubSample Correlation Matrix \n (use for reference)", fontsize=14)
```

Corelation Conclusion :
1. Positive corelation - v2, v4, v11, v19 seem to be positively coreleated with the feature class.
   Higher the positive corelation value, higher the possibility of the transaction being fraud.
2. Negative corelation - v10, v12, v14, v17 seem to be negatively coreleated with the feature class.
   Lower the negative corelation value, higher the possibility of the transaction being fraud.

```python
f, axes = plt.subplots(ncols=4, figsize=(20,4))
colors = ['yellow', 'orange']

# Negative Correlations with our Class (The lower our feature value the more likely it will be a fraud transaction)
sns.boxplot(x="Class", y="V17", data=df, palette=colors, ax=axes[0])
axes[0].set_title('V17 vs Class Negative Correlation')

sns.boxplot(x="Class", y="V14", data=df, palette=colors, ax=axes[1])
axes[1].set_title('V14 vs Class Negative Correlation')


sns.boxplot(x="Class", y="V12", data=df, palette=colors, ax=axes[2])
axes[2].set_title('V12 vs Class Negative Correlation')


sns.boxplot(x="Class", y="V10", data=df, palette=colors, ax=axes[3])
axes[3].set_title('V10 vs Class Negative Correlation')
```

```python
f, axes = plt.subplots(ncols=4, figsize=(20,4))
colors = ['yellow', 'orange']

# Positive correlations (The higher the feature the probability increases that it will be a fraud transaction)
sns.boxplot(x="Class", y="V11", data=df, palette=colors, ax=axes[0])
axes[0].set_title('V11 vs Class Positive Correlation')

sns.boxplot(x="Class", y="V4", data=df, palette=colors, ax=axes[1])
axes[1].set_title('V4 vs Class Positive Correlation')


sns.boxplot(x="Class", y="V2", data=df, palette=colors, ax=axes[2])
axes[2].set_title('V2 vs Class Positive Correlation')


sns.boxplot(x="Class", y="V19", data=df, palette=colors, ax=axes[3])
axes[3].set_title('V19 vs Class Positive Correlation')
plt.show()
```

We can notice that the positive and negative correlation features identified do co-relate as we concluded form the heatmap
